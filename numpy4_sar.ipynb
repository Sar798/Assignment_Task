### Statistical Measures on SepalWidth

```python
sepal_width
```

```python
sepal_width.mean()
```

```python
np.median(sepal_width)
```

```python
range = np.max(sepal_width)-np.min(sepal_width)
range
```

```python
q1 = np.quantile(sepal_width,0.25)
q1
```

```python
q3 = np.quantile(sepal_width,0.75)
q3
```

```python
IQR = q3-q1
IQR
```

```python
lower = q1 - (1.5*IQR)
lower
```

```python
upper = q3 + (1.5*IQR)
upper
```

```python
sepal_width[sepal_width<lower]
```

```python
sepal_width[sepal_width<upper]
```

```python
np.var(sepal_width)
```

```python
np.std(sepal_width)
```

### Statistical Measures on SepalLength
