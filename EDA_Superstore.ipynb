### Outlier

```python
df3 = df.select_dtypes(exclude="object")
for i in df3.columns:
    Q1 = df3[i].quantile(0.25)
    Q3 = df3[i].quantile(0.75)

    IQR = Q3-Q1
    lower = Q1 - (1.5*IQR)
    upper = Q3 + (1.5*IQR)
    df3[(df3[i]<lower)] = lower
    df3[(df3[i]>upper)]= upper
    print(f"  {i} Lower_limit {lower} ")
    print(f"  {i} Upper_limit {upper} ")
    print()
```
